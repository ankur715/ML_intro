## SP01:

```python
#SP01 Files:
SP01a = pd.read_csv(CSVpath+"01_SP01a.csv")
SP01b = pd.read_csv(CSVpath+"02_SP01b.csv")
SP01c = pd.read_csv(CSVpath+"03_SP01c.csv")
SP01d = pd.read_csv(CSVpath+"04_SP01d.csv")
SP01e = pd.read_csv(CSVpath+"05_SP01e.csv")
SP01f = pd.read_csv(CSVpath+"06_SP01f.csv")
SP01g = pd.read_csv(CSVpath+"07_SP01g.csv")
SP01h = pd.read_csv(CSVpath+"08_SP01h.csv")
SP01i = pd.read_csv(CSVpath+"09_SP01i.csv")
SP01j = pd.read_csv(CSVpath+"10_SP01j.csv")
SP01l = pd.read_csv(CSVpath+"12_SP01l.csv")
SP01m = pd.read_csv(CSVpath+"13_SP01m.csv")
```

```python
#concatenate SP01 files together
SP01 = pd.concat([SP01a,SP01b,SP01c,SP01d,SP01e,SP01f,
                 SP01g,SP01h,SP01i,SP01j,SP01l,SP01m])
SP01.head()
```

```python
#SP01 Mapping sheets
SP01_str = pd.read_excel(path+"SP01 Mapping.xlsx", sheet_name="Structure Mapping")
SP01_form = pd.read_excel(path+"SP01 Mapping.xlsx", sheet_name="Form Mapping")
SP01_chl = pd.read_excel(path+"SP01 Mapping.xlsx", sheet_name="Channel Mapping")
SP01_pyrgrp = pd.read_excel(path+"SP01 Mapping.xlsx", sheet_name="Payer Group Mapping")
SP01_pyr = pd.read_excel(path+"SP01 Mapping.xlsx", sheet_name="Payer Mapping")
SP01_unit = pd.read_excel(path+"SP01 Mapping.xlsx", sheet_name="Unit Conversion")
```

### SP01 - Structure Mapping:

```python
SP01 = pd.merge(SP01, SP01_str, on=["PRODUCT_NAME_SP","NDC_CODE"], how="outer")
SP01.head()
# https://pythonprogramming.net/join-merge-data-analysis-python-pandas-tutorial/
```

### SP01 - Form Mapping:

```python
SP01_form.columns = ["PRODUCT_NAME_SP", "NDC_CODE", "NNI Form"]
SP01 = pd.merge(SP01, SP01_form, on=["PRODUCT_NAME_SP","NDC_CODE"], how="outer")
SP01.head()
```

### SP01 - Channel Mapping:

```python
SP01 = pd.merge(SP01, SP01_chl.iloc[:,:2], on="CHANNEL", how="outer")
SP01.head()
```

### SP01 - Payer Group Mapping:

```python
SP01_pyrgrp1 = SP01_pyrgrp.iloc[:,:4]
SP01_pyrgrp1.columns = ["CHANNEL", "CUSTOMER_GROUP", "PLAN_NAME", "NNI Payer Group"]
SP01 = pd.merge(SP01, SP01_pyrgrp1, on=["CHANNEL", "CUSTOMER_GROUP", "PLAN_NAME"], how="outer")
SP01.head()
```

### SP01 - Payer Mapping:

```python
SP01 = pd.merge(SP01, SP01_pyr.iloc[:,:3], on=["NNI Channel", "NNI Payer Group"], how="outer")
SP01.head()
```

### SP01 - Unit Conversion Mapping:

```python
SP01 = pd.merge(SP01, SP01_unit, on=["NNI Product", "NNI Form"], how="outer")
SP01.head()
```

```python
SP01["Converted Units"] = SP01["SUBMITTED_UNITS"] * SP01["Conversion Factor"]
SP01.head()
```

```python
SP01["Conversion Factor"].unique()
```

```python
SP01["Converted Units"].unique()
```

```python
len(SP01)
```

```python
#Save DataFrame as new CSV
SP01.to_csv("PCFM Mapping Python.csv")
```
