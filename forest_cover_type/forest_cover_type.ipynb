### Tuning using GridSearchCV

```python
# these param grids were found online with the most similar data

param_grid_log = {"C": [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
param_grid_lsvc = {"C": [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
param_grid_nc = {"shrink_threshold": [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
param_grid_dt = {"max_depth" : [1, 5, 10, 15, 20, 25, 30],
                 "min_samples_leaf" : [1, 2, 4, 6, 8, 10]}
param_grid_rf = {'max_depth': [80, 90, 100, 110],
                'max_features': [2, 3],
                'min_samples_leaf': [3, 4, 5],
                'min_samples_split': [8, 10, 12],
                'n_estimators': [100, 200, 300, 1000]}

param_list = []
param_list.append(param_grid_log)
param_list.append(param_grid_lsvc)
param_list.append(param_grid_nc)
param_list.append(param_grid_dt)
param_list.append(param_grid_rf)
```

```python
models_tuned = []

skfold = model_selection.StratifiedKFold(n_splits=10, random_state=7)
plt.figure()
for i, classifier in enumerate(classifiers):
    print("RUNNING FOR {}".format(classifier_namess[i]))
    l = classifier
    gc = GridSearchCV(l, param_list[i], cv=skfold, scoring='r2', n_jobs=-1)
    gc.fit(X1_train_sc, y1_train)

    print("Best cross-validation score: {:.2f}".format(gc.best_score_))
    print("Best parameters: ", gc.best_params_)
    gc.best_estimator_
    models_tuned.append(gc.best_estimator_)
    
    train_scores_mean = gc.cv_results_["mean_train_score"]
    train_scores_std = gc.cv_results_["std_train_score"]
    test_scores_mean = gc.cv_results_["mean_test_score"]
    test_scores_std = gc.cv_results_["std_test_score"]

    alphas = np.logspace(2, 10, num=len(train_scores_mean), base=2)

    plt.subplot(2, 3, i+1)
    plt.title(classifier_namess[i])
    plt.xlabel("$\\alpha$ (alpha)")
    plt.ylabel("Score")
    # plot train scores
    plt.semilogx(alphas, train_scores_mean, label="Mean Train score", color="navy")

    # create a shaded area between [mean - std, mean + std]
    plt.gca().fill_between(alphas, train_scores_mean - train_scores_std,
                           train_scores_mean + train_scores_std, alpha=0.2, color="navy")
    plt.semilogx(alphas, test_scores_mean, label="Mean Test score", color="darkorange")

    # shaded graph of MSE
    plt.gca().fill_between(alphas, test_scores_mean - test_scores_std,
                           test_scores_mean + test_scores_std, alpha=0.2, color="darkorange")
    plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
