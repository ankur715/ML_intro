```python
sns.set(rc={'figure.figsize':(15,15)})
sns.heatmap(df.corr(),annot=True,cmap='Reds')
plt.show()
```

```python
# Barplot of Month vs Special Day
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(df['Month'], df['SpecialDay'], color='green')
plt.show()
# Feb Valentine's Day, May Mother's Day
```

```python
# Count plot of Revenue
sns.set(rc={'figure.figsize':(5,5)})
sns.countplot(df['Revenue'])
plt.show()
# 1/6 people bought product
```

```python
# Violin plot of Revenue vs Product related duration¶
sns.set(rc={'figure.figsize':(5,5)})
sns.violinplot(x="Revenue", y="ProductRelated_Duration", data=df)
plt.show()
```

```python
# Multivariate Analysis¶
sns.set(rc={'figure.figsize':(15,15)})
sns.catplot(x="Revenue",y="ProductRelated_Duration",hue = "Weekend",col="Region",data=df,col_wrap=3,height=2.5, aspect=.8)
plt.show()
# summarizes the distribution of Revenue vs Time spent on the website to buy a product for each region, 
# taking into consideration if its a weekend or not
```
